```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import requests
import random
import time

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
```

```python
service = Service('C:\\Users\\Nizar Faris\\Lamudi\\chromedriver\\chromedriver.exe')
driver = webdriver.Chrome(service=service)
```

```python
#Create empty list for append data from scrapping
Price = []
Installment = []
Description = []
Location = []
Bedroom = []
Bathroom = []
Carport = []
Land = []
Building = []
text = []
info = []

#Create empty dataframe for extend data after append
df = pd.DataFrame({
    'Price': [], 
    'Installment': [], 
    'Description': [],
    'Location': [],
    'Bedroom': [], 
    'Bathroom': [], 
    'Carport': [],
    'Land': [], 
    'Building': [], 
})
df
```

```python

base_url = 'https://www.rumah123.com/jual/rumah/?bathroom=1&bedroom=1&certificates%5B%5D=1&minBuiltupSize=1&minLandArea=1&q=bandung&page='

pagenum = 1
maxpage = 1000
while pagenum < maxpage:
    try:
        url = f'{base_url}{pagenum}'
        driver.get(url)

        #to avoid captcha and mimic human behaviour
        wait = WebDriverWait(driver, 4)
        time.sleep(random.uniform(2, 7))

        my_page = driver.page_source
        my_html = BeautifulSoup(my_page, 'html.parser')
        print(f'Scrapping on page {pagenum}')
        pagenum += 1

        try:
            container = my_html.findAll('div', {'class': 'ui-organism-intersection__element intersection-card-container'})
            for i in range(len(container)):
                try: #Price
                    price = container[i].find('div', class_='card-featured__middle-section__price').find('strong').text
                    Price.append(price)
                except:
                    Price.append('No Data')
    
                try: #Installment
                    instal = container[i].find('button', class_='card-featured__middle-section__installment').find('span').text
                    Installment.append(instal)
                except:
                    Installment.append('No Data')
                
                try: #Description
                    desc = container[i].find('a').find('h2').text
                    Description.append(desc)
                except:
                    Description.append('No Data')
                
                try: #Location
                    loc = container[i].find('span', string=lambda text: text.endswith(' Bandung')).text
                    Location.append(loc)
                except:
                    Location.append('No Data')
    
                try:
                    attribute_house = container[i].find('div', class_='card-featured__middle-section__attribute').findAll('span', class_='attribute-text')
                    attributes = []
                
                    for index, attribute in enumerate(attribute_house):
                        if index < 3:
                            attributes.append(attribute.text)
                        else:
                            break
                
                    while len(attributes) < 3:
                        attributes.append('0')
                
                    bedroom, bathroom, carport = attributes[:3]
                    Bedroom.append(bedroom)
                    Bathroom.append(bathroom)
                    Carport.append(carport)
                
                except:
                    bedrooms.append('No Data')
                    bathrooms.append('No Data')
                    carports.append('No Data')
                
                try:
                    attribute_area = container[i].find('div', class_='card-featured__middle-section__attribute').findAll('div', class_='attribute-info')
                    attributes = []
                
                    for index, attribute in enumerate(attribute_area):
                        if index < 2:
                            attributes.append(attribute.text)
                        else:
                            break
                
                    while len(attributes) < 2:
                        attributes.append('0')
                
                    land, building = attributes[:2]
                    Land.append(land)
                    Building.append(building)
                except:
                    Land.append('No Data')
                    Building.append('No Data')

        except Exception as e:
            print(f'Error scraping in page {pagenum}: {e}')
            break
        
    except Exception as e:
        print(f'Scrapping failed: {e}')
        break


df_temp = pd.DataFrame({
    'Price': pd.Series(Price), 
    'Installment': pd.Series(Installment), 
    'Description': pd.Series(Description),
    'Location': pd.Series(Location),
    'Bedroom': pd.Series(Bedroom), 
    'Bathroom': pd.Series(Bathroom),
    'Carport': pd.Series(Carport),
    'Land': pd.Series(Land), 
    'Building': pd.Series(Building)
})

df = pd.concat([df, df_temp], axis=0)

driver.quit()

print('Done scraping all locations')
df
```
